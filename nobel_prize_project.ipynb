Kimya, Edebiyat, Barış, Fizik ve Tıp kategorilerindeki 2000 sonrasındaki kişilerin ülkelerini, yaşlarını görselleştirin.(Her bir Kategori için ayrı görselleştirme yapılması istenmektedir) Veriyi yorumlayınız.

```python


# 2000 yılı sonrası ödül kazananları filtreleme
df['year'] = pd.to_datetime(df['year'], format='%Y')
df_2000_onwards = df[df['year'].dt.year > 2000]

# İlgili kategorileri seçme
categories = ['Chemistry', 'Literature', 'Peace', 'Physics', 'Physiology or Medicine']
df_filtered = df_2000_onwards[df_2000_onwards['category'].isin(categories)]

# Yaş hesaplama (doğum tarihine göre)
df_filtered['birth_date'] = pd.to_datetime(df_filtered['birth_date'], errors='coerce')
df_filtered['age'] = df_filtered['year'].dt.year - df_filtered['birth_date'].dt.year

# Boş verileri kontrol etme ve kaldırma
df_filtered = df_filtered.dropna(subset=['birth_country', 'age'])

# Her bir kategori için görselleştirme
for category in categories:
    df_category = df_filtered[df_filtered['category'] == category]
    
    if df_category.empty:
        print(f"No data available for category: {category}")
        continue
    
    # Ülkelere göre ödül kazananların sayısı
    plt.figure(figsize=(10, 6))
    df_category['birth_country'].value_counts().plot(kind='bar')
    plt.title(f'Number of Nobel Prize Winners by Country in {category} (2000 onwards)')
    plt.xlabel('Country')
    plt.ylabel('Number of Winners')
    plt.show()
    
    # Yaş dağılımı
    plt.figure(figsize=(10, 6))
    df_category['age'].plot(kind='hist', bins=20)
    plt.title(f'Age Distribution of Nobel Prize Winners in {category} (2000 onwards)')
    plt.xlabel('Age')
    plt.ylabel('Frequency')
    plt.show()
```
